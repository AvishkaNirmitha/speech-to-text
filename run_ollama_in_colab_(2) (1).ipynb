```python
import threading
import time
import GPUtil

def monitor_gpu_usage(interval=1):
    """
    Function to monitor GPU memory usage at regular intervals.
    """
    while monitor_flag:
        gpus = GPUtil.getGPUs()
        for gpu in gpus:
            print(f"GPU {gpu.id}: {gpu.memoryUsed}MB / {gpu.memoryTotal}MB")
        time.sleep(interval)



# Set up GPU monitoring
monitor_flag = True
monitor_thread = threading.Thread(target=monitor_gpu_usage, args=(1,))
monitor_thread.start()

# Execute the ask_question function
try:
    question = "What the hell ?"
    ask_question(question)
finally:
    # Stop the GPU monitoring thread after the function execution
    monitor_flag = False
    monitor_thread.join()
```
